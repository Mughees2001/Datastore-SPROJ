```python
from server import Server
from client import Client
from lib.log import logger
from threading import Thread
import time
```

```python
def OnServer(server_instance):
    server_thread = Thread(target=start_server, args=(server_instance,))
    server_thread.daemon = True
    server_thread.start()

def start_server(server_instance):
    try:
        server_instance.run()
    except Exception as e:
        logger.error(f"Error starting server: {e}")
```

```python
src_port = 10013
dst_port = 10014
src_server = Server('localhost',str(src_port))
dst_server = Server('localhost',str(dst_port))
OnServer(src_server)
OnServer(dst_server)
# Ensure the server has time to start up
time.sleep(1)
```

```python
client = Client("localhost",str(src_port))  # Connect to the server here
```

```python
for i in range(100):
    client.put(str(i), "value".encode())
```
